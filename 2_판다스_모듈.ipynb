```python
# 분산, 표준편차는 데이터가 평균으로부터 얼마나 떨어져 있는지 정도
# 분산 : (데이터 - 평균)**2을 모두 합한 값 / 데이터의 개수
# 표준편차 : 분산에 루트 씌움

import numpy as np

data1 = np.array([1, 3, 5, 7, 9])
data2 = np.array([3, 4, 5, 6, 7])

print(((1-5)**2 + (3-5)**2 + (5-5)**2 + (7-5)**2 + (9-5)**2) / 5) # 분산
print(np.sqrt(((1-5)**2 + (3-5)**2 + (5-5)**2 + (7-5)**2 + (9-5)**2) / 5)) # 표준편차
```

```python
print(data1.var()) # 분산
print(data1.std()) # 표준편차
```

```python
print(data2.var()) # 분산
print(data2.std()) # 표준편차
```
