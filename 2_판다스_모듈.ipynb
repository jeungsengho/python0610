### **4-6. 정렬하기**

```python
# index로 오름차순 정렬
df.sort_index()
```

```python
# index로 내림차순 정렬
df.sort_index(ascending=False)
```

```python
# 값에 따른 정렬
df.sort_values(by='height')   # 오름차순
```

```python
df.sort_values(by='height', ascending=False)  # 내림차순
```

```python
df.sort_values(by='height', ascending=False, na_position='first')   # NaN을 가장 위로
```

```python
# 2차 정렬 : 키(내림차순), 브랜드 평판지수(내림차순)

df.sort_values(by=['height', 'brand'], ascending=[False, True], na_position='first') 
```

### **5-1. 범위선택**

```python
df.loc[:3, 'name']              # loc 3 까지(포함) ,iloc 3 미만(미포함)
```

```python
df.loc[:, ['name', 'birthday']]
```

```python
df.loc[3:8, ['name', 'birthday']]                    # , = 개별
```

```python
df.loc[3:8, 'name':'birthday']                        # : = 범위
```

```python
df.iloc[:, [0, 2]]
```
