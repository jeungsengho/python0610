* [민간 아파트 가격동향](https://bit.ly/ds-house-price)


```python
import pandas as pd
```

```python
df = pd.read_csv('https://bit.ly/ds-house-price')
```

```python
df
```

```python
df.info()
```

```python
df = df.rename(columns={'분양가격(㎡)':'분양가격'})
```

```python
df
```

```python
df.describe()
```

```python
# 분양가격을 int로 변환
df['분양가격'].astype(int)    # ValueError: invalid literal for int() with base 10: '  ' -> 공백존재해서
```

```python
# '   '  공백이 존재하는 데이터가 있는지 확인
df.loc[df['분양가격'] == '  ']
```

```python
# strip()를 사용하여 공백이 있는 데이터의 공백을 삭제
df['분양가격'] = df['분양가격'].str.strip()
```

```python
df.loc[df['분양가격'] == '  ']
```

```python
df['분양가격'].astype(int)  # ValueError: invalid literal for int() with base 10: '' -> 데이터가 없어서
```

```python
df.loc[df['분양가격'] == '']  # 데이터가 없는 경우
```

```python
# 데이터가 없는 경우는 0을 채워줌
df.loc[df['분양가격'] == '', '분양가격'] = 0
```

```python
df.loc[df['분양가격'] == '']
```

```python
df['분양가격'].astype(int)  # ValueError: cannot convert float NaN to integer -> NaN값 때문
```

```python
# NaN을 0으로 변환
df['분양가격'] = df['분양가격'].fillna(0)
```

```python
df['분양가격'].astype(int)  # ValueError: invalid literal for int() with base 10: '6,657' -> ,가 들어있어서 정수형 변환 에러
```

```python
# 콤마를 제거
df['분양가격'] = df['분양가격'].str.replace(',', '')
```

```python
df['분양가격'].astype(int)  # ValueError: cannot convert float NaN to integer -> NaN 때문에
```

```python
# NaN을 0으로 변환
df['분양가격'] = df['분양가격'].fillna(0)
```

```python
df['분양가격'].astype(int)  # ValueError: invalid literal for int() with base 10: '-'
```

```python
# '-'을 제거
df['분양가격'] = df['분양가격'].str.replace('-', '')
```

```python
df['분양가격'].astype(int)  # ValueError: cannot convert float NaN to integer
```

```python
# NaN을 0으로 변환
df['분양가격'] = df['분양가격'].fillna(0)
```

```python
df['분양가격'].astype(int)  # ValueError: invalid literal for int() with base 10: ''
```

```python
# 데이터가 없는 경우는 0을 채워줌
df.loc[df['분양가격'] == '', '분양가격'] = 0
```
